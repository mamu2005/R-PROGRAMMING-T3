```python
import pandas as pd
```

```python
df=pd.read_csv('/content/train_revised.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Extract the relevant columns
X = df['ride_id'].values.reshape(-1, 1)

# Convert travel_time to minutes
df['travel_time'] = pd.to_datetime(df['travel_time'], format='%H:%M').dt.hour * 60 + pd.to_datetime(df['travel_time'], format='%H:%M').dt.minute
y = df['travel_time']

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the values
y_pred = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('ride_id')
plt.ylabel('travel_time')
plt.title('Linear Regression of ride_id vs travel_time')
plt.show()
```

```python
# prompt: draw a linear regression graph betwen travel_date and travel_time

import pandas as pd
import matplotlib.pyplot as plt
# Convert travel_date to numerical values (e.g., using ordinal dates)
df['travel_date_ordinal'] = pd.to_datetime(df['travel_date']).apply(lambda x: x.toordinal())

# Extract the relevant columns
X = df['travel_date_ordinal'].values.reshape(-1, 1)
y = df['travel_time']

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the values
y_pred = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('travel_date_ordinal')
plt.ylabel('travel_time')
plt.title('Linear Regression of travel_date vs travel_time')
plt.show()
```
